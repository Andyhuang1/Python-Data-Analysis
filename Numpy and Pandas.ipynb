## 5. 广播

广播可以简单理解为用于不同大小数组的二元通用函数（加、减、乘等）的一组规则，同之前的常用通用函数一样，都是通过向量化的操作减少缓慢的python循环

### 5.1 广播介绍

```python
# 对于同样大小的数组，二元运算符是对相应元素逐个计算
a = np.array([0, 1, 2])
b = np.array([5, 5, 5])
print(a + b)
# 广播允许二元运算符用于不同大小的数组，例如简单将一个标量（可以理解为一个零维的数组）和一个数组相加
# 可以理解为先将数值5扩展为数组[5, 5, 5]然后执行加法
print(a + 5)
```

```python
# 将一维数组array([0, 1, 2])扩展成匹配3x3的二维数组array([[0, 1, 2],[0, 1, 2],[0, 1, 2]])
M = np.ones((3, 3))
print(M)
print(a)
M + a
```

```python
# 两个数组同时广播
a = np.arange(3)
b = np.arange(3)[:, np.newaxis]
print(a)
print(b)
a + b
```

### 5.2 广播规则

规则1：如果两个数组的维度数不同，那么小维度数组的形状将会在最左边补1

规则2：如果两个数组的形状在任何一个维度上都不匹配，那么数组的维度会沿着维度为1的维度扩展以匹配另外一个数组的形状

规则3：如果两个数组的形状在任何一个维度上都不匹配并且没有任何一个维度等于1，那么就会引发异常

```python
# 广播示例 1 ：将一个二维数组和一个一维数组相加
M = np.ones((2, 3))
a = np.arange(3)
print(M, a)
print(M.shape, a.shape)
print('根据规则1，数组a的维度更小，所以在其左边补1，m.shape(2, 3) a.shape(1, 3)')
print('根据规则2，两个数组第一个维度不匹配，扩展数组a以匹配数组，a.shape(2, 3)')
M + a
```

```python
# 广播示例 2：两个数组均需要广播
a = np.arange(3).reshape(3, 1)
b = np.arange(3)
print(a, b)
print(a.shape, b.shape)
print('根据规则1，数组b的维度更小，所以在其左边补1，a.shape(3, 1) b.shape(1, 3)')
print('根据规则2，两个数组维度不匹配，扩展两个数组，a.shape(3, 3) b.shape(3, 3)')
print('''
[[0, 0, 0],
 [1, 1, 1],
 [2, 2, 2]] + [[0, 1, 2],
               [0, 1, 2],
               [0, 1, 2]] = \n''', a + b)
```

### 5.3 广播实际应用

1. 数组的归一化

数组归一化是将数组的数值范围缩放到特定范围或标准化的过程。在数据处理和机器学习中，常常需要对数据进行归一化，以便使数据具有一致的尺度和范围，从而有助于模型的训练和预测。

常见的数组归一化方法包括：

1. 最小-最大归一化（Min-Max Scaling）： 这是最常见的归一化方法之一。它将数据缩放到指定的范围，通常是 [0, 1]。

2. 标准化（Standardization）： 这种方法将数据转换为均值为 0、标准差为 1 的正态分布（也叫标准正态分布）。标准化使得数据分布更符合统计假设，有助于某些机器学习算法的性能。

为什么要进行数组归一化？

1. 消除尺度影响： 不同的特征可能具有不同的尺度和范围，这可能会对某些机器学习算法产生负面影响。通过归一化，可以消除特征之间的尺度差异，使得算法更加稳定。

2. 加速收敛： 对于许多优化算法（如梯度下降），归一化可以加速算法的收敛过程，使其更快地达到最优解。

3. 提高模型性能： 在某些情况下，归一化可以提高机器学习模型的性能，因为模型更容易学习和泛化归一化后的数据。

4. 可视化和解释性： 归一化后的数据更容易可视化和解释，因为它们具有相似的尺度。

需要注意的是，在某些情况下，不需要进行数组归一化，特别是当数据已经在合适的尺度范围内时，或者某些特定的机器学习算法不敏感于尺度差异时。因此，在进行数组归一化之前，需要仔细考虑数据的性质和算法的要求。

rand.multivariate_normal: 这个函数生成多元正态分布的随机数，也被称为多维高斯分布。生成的随机数是多维的，通常用于生成多维数据，如多维特征的数据集。接受均值向量和协方差矩阵作为参数，用于定义多维正态分布的性质。成的随机数是一个二维数组，其中每一行代表一个多维随机数据点。

np.random.normal: 这个函数生成单变量正态分布（univariate normal distribution）的随机数，也被称为高斯分布。生成的随机数是单个值，通常用于生成一维数据或单个随机变量的值。接受均值和标准差作为参数，用于定义单变量正态分布的性质。生成的随机数是一个一维数组，包含单变量随机数值。

### 7.5 示例：数据区间划分，手动创建直方图

```python
# 用ufun.c.at,假定有1000个值，快速统计分布在各个区间中的数据频次
np.random.seed(42)
x = np.random.randn(100)

# 手动计算直方图，生成一个[-5, 5]之间均匀生成的20个数据点，划分成20个均匀的小区间
bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins) # 生成一个和bins数组形状一样的全零数组
print('bins:', bins)
print('counts:', counts)

# 为每个x找到合适的区间
"""
np.searchsorted(bins, x): 这是调用NumPy中的np.searchsorted函数，该函数接受两个参数：
第一个参数 bins 是已排序的数组，表示被查找的范围。
第二个参数 x 是要查找的数值。
np.searchsorted函数返回的结果是一个整数，表示在将 x 插入到 bins 数组中时，应该插入的位置。具体来说：
如果 x 小于 bins 中的所有值，那么返回值将是0，表示 x 应该插入到 bins 数组的开头。
如果 x 大于 bins 中的所有值，那么返回值将是 len(bins)，表示 x 应该插入到 bins 数组的末尾。
如果 x 介于 bins 中的两个值之间，那么返回值将是介于它们之间的位置，以保持 bins 的升序顺序。
"""
i = np.searchsorted(bins, x)

# 为每个区间加上1, 计数数组counts反映的是在各个区间中的点的个数
np.add.at(counts, i, 1)

# 画出结果
plt.plot(bins, counts, linestyle = '-')
plt.show()

# 用plt.hist()一步实现上述效果
plt.hist(x, bins, histtype = 'step')
```

## 8. 数组的排序

```python
# 简单的选择排序重复寻找列表中的最小值，并且不断交换直到列表是有序的
def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
        return x
    
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
```

### 8.3 示例：K个最近邻

以下示例展示如何使用argsort函数沿着多个轴快速找到集合中每个点的最近邻。

```python
# 首先在二维平面上创建一个有10个随机点的集合，按照惯例这些数据点放在一个10X2的数组中
X = rand.rand(10, 2)
print('X: \n', X)

# 将这些点画散点图
import matplotlib.pyplot as plt
import seaborn; seaborn.set # 设置绘图风格

plt.scatter(X[:, 0], X[:, 1], s = 100)
```

现在来计算两两数据点对间的距离，两点间的距离的平方等于每个维度距离差的平方的和，利用广播和聚合功能计算矩阵的平方距离。

```python
dist_sq = np.sum((X[:,np.newaxis,:] - X[np.newaxis,:,:]) **2, axis = -1)
dist_sq.shape
```

```python
# 确认以上步骤，应该看到该矩阵的对角线（也就是每个点到其自身的距离）都是0
dist_sq.diagonal()
```

```python
# 得到一个转化为两点间的平方距离的矩阵后，就可以使用np.argsort函数沿着每行进行排序，最左边的列给出的索引值就是最近邻
# 需要注意的是，第一列是按照0-9从小到大顺序排列的，这是因为每个点的最近邻是其自身
nearest = np.argsort(dist_sq, axis = 1)
nearest
```

如果使用全排序，我们实际上可以实现的比这个例子展示的更多。如果仅仅关心k个最近邻，那么唯一需要做的是分隔每一行，这样最小的k + 1的平方距离将排在最前面，其他更长距离占据矩阵该行的其他位置。可以用np.argpartition函数实现：

```python
k = 2
nearest_partition = np.argpartition(dist_sq, k + 1, axis = 1)
nearest_partition
```

```python
# 为了将邻节点网络可视化，我们将每个点和其最近的两个最近邻连接
plt.scatter(X[:, 0], X[:, 1], s = 100)

k = 2
for i in range(X.shape[0]):
    for j in nearest_partition[i, :k+1]:
        # 画一条从X[i]到X[j]的线段
        # 用zip方法实现
        plt.plot(*zip(X[j], X[i]), color = 'black')
```

一种方法是通过一个覆盖全局的掩码表示缺失值，另一种方法是用一个标签值表示缺失值

### 5.2 Pandas的缺失值

Pandas选择用标签方法表示缺失值，包括两种Python原有的缺失值：浮点数据类型的NaN值，以及Python的None对象

```python
# 1.None：Python对象类型的缺失值，只能用于'object'数组类型，即由python对象构成的数组
vals1 = np.array([1, None, 3, 4])
vals1
```

```python
for dtype in ['object', 'int']:
    print('dtype = ', dtype)
    %timeit np.arange(1E6, dtype = dtype).sum()
    print()
```

这里dtype = object表示Numpy认为这个数组是Python对象构成的，因此其类型判断为object,这种类型比其他原生类型数组要消耗更多资源。

Python中没有定义整数与None之间的加法预算，因此对使用Python对象构成的包含None的数组进行累计操作sum()或min()等会报错

```python
# 2.NaN:数值类型的缺失值，全称Not a Number,在任何系统都兼容的特殊浮点数
vals2 = np.array([1, np.nan, 3, 4])
print(vals2.dtype)
# 无论和NaN进行任何操作，最终结果都是NaN
print('1 + np.nan :', 1 + np.nan)
print('0 * np.nan :', 0 * np.nan)
print(vals2.sum(), vals2.min(), vals2.max())
# 使用Numpy特殊的累计函数可以忽略缺失值的影响
print(np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2))
```

```python
# Pandas中NaN与None的差异：Pandas看作等价交换，会将没有标签的数据类型,或None自动转换为NaN
print(pd.Series([1, np.nan, 2, None]))
x = pd.Series(range(2), dtype = int)
print('x: \n', x)
x[0] = None
x
```

Pandas对不同类型缺失值的转换规则:

|类型                 | 缺失值转换规则               | NA 标签值                     |
|--------------       |-----------------------------|------------------------|

| ``object对象类型``  | 无变化                   | ``None`` or ``np.nan`` |
| ``integer整数类型`` | 强制转换为 ``float64``         | ``np.nan``             |

## 6. 层级索引

Series和DataFrame主要是对一维和二维数据储存，对于多维数据，Pandas提供了Panel和Panel4D对象解决三维数据和四维数据；但在实践中，更直观的形式是通过层级索引（多级索引）配合多个有不同等级的一级索引一起使用，将多维数组转化为类似一维Series和二维DataFrame的形式

### 6.1 多级索引Series

如何用一维的Series对象表示二维数据：用一系列包含特征与取值的数据点进行简单演示

```python
# 1.笨方法
# 假设想要分析美国各州在不同年份的数据，可以用python元组来表示索引
index = [('加州', 2000), ('加州', 2010), ('纽约', 2000), ('纽约', 2010), ('新泽西', 2000), ('新泽西', 2010)]
populations = [33871648, 37253956, 18986457, 19378102, 20851820, 25145561]
pop = pd.Series(populations, index = index)
print(pop)
# 通过元组构成的多级索引，直接在Series上取值和用切片查询数据
print('取加州2000至新泽西2000的数据')
print(pop[('加州', 2000):('新泽西', 2000)])
# 这样做很不方便，假如要取所有2010年的数据，需要用循环找出2010年
for i in pop.index:
    if i[1] == 2010:
        print(i)

pop[[i for i in pop.index if i[1] == 2010]] # 方法不够简洁，处理较大数据时不够高效
```

```python
# 好方法：Pandas多级索引，MultiIndex类型提供了更丰富的方法，用元组创建一个多级索引
index = pd.MultiIndex.from_tuples(index)
# levels是一个列表，元素为两个子列表，一个列表包含州名，一个列表包含年份
print(index)
print('将pop的索引重置为MultiIndex，就会看到层级索引')
pop  = pop.reindex(index)
print(pop) 
# 前两列表示Series的多级索引值，第三列是数据，可以直接用第二个索引获得2010年的数据，与Pandas的切片查询用法一致
pop[:, 2010] # 结果是一个单索引的数组，与笨方法相比语法更简单操作更方便
```

```python
# 高维数据的多级索引,unstack()快速将一个多级索引Series转化为普通索引的DataFrame, stack()相反
print(pop)
print('将一个多级索引的Series转化为普通索引的DataFrame')
print(pop.unstack())
print('将一个普通索引的DataFrame转化为多级索引的Series')
print(pop.unstack().stack())
# 能用一个多级索引的Series表示二维数据能用一个多级索引的Series或者DataFrame表示三维或更高维数据
pop_df = pd.DataFrame({'total':pop, 'under18': [9267089, 9284094, 4687374, 4318033, 5906301, 6879014]})
pop_df
```

```python
# Pandas数据运算方法介绍过的通用函数和其他功能同样适用于层级索引，例如计算18岁以下的人口占总人口的比例
f_u18 = pop_df['under18'] / pop_df['total']
f_u18.unstack()
```

```python
# 定义一个能够创建DataFrame某种形式的函数
def make_df(cols, ind):
    """一个简单的DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

# DataFrame示例
make_df('ABC', range(3))
```

### 7.2 通过pd.concat实现简易合并

```python
# pd.concat(
#     objs: 'Iterable[NDFrame] | Mapping[HashableT, NDFrame]',
#     *,
#     axis: 'Axis' = 0,
#     join: 'str' = 'outer',
#     ignore_index: 'bool' = False, 设置True会忽略索引合并后创建一个新的整数索引
#     keys=None, 设置多级索引标签
#     levels=None,
#     names=None,
#     verify_integrity: 'bool' = False, 设置True时有重复索引会报错
#     sort: 'bool' = False,
#     copy: 'bool | None' = None,
# ) -> 'DataFrame | Series'

# pd.concat()可以简单的合并一维的Series或DataFrame对象，与np.concatenate()合并数组一样
ser1 = pd.Series(['A', 'B', 'C'], index = [1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index = [4, 5, 6])
print(pd.concat([ser1, ser2]))

# 合并高维数据
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])
print(df1)
print(df2)
print(pd.concat([df1, df2]))

# pd.concat默认axis = 0即沿着行方向合并，也可以设置沿着列方向拼接
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
print(df3)
print(df4)
print(pd.concat([df3, df4]))
pd.concat([df3, df4], axis = 1)
```

### 8.2 数据连接的类型

pd.merge()函数实现了三种数据连接的类型：一对一、多对一和多对多

```python
import pandas as pd
import numpy as np

# 设置展现display输出并排显示结果
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
# 1.一对一连接，按列合并，类似数据库的innerjoin自动以相同的列作为键进行连接, 默认丢弃原来的行索引
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
df3 = pd.merge(df1, df2)
display('df1', 'df2', 'df3')
```

### 9.3 GroupBy:分割、应用和组合

对某些标签或索引的局部进行累计分析需要用到groupby, 虽然分组(group by)这个名字借用SQL，其理念引用发明R语言frame的Hadley Wickkam的观点更适合：

分割（split）、应用（apply）、组合（combine）

#### 1. 分割、应用和组合

下图清晰描述了GroupBy的过程

-分割：将DataFrame按照指定的键key分割成若干组

-应用：每个组应用函数，通常是累计、转换或过滤函数

-组合：将每一组的结果合并成一个输出数组

中间分割过程不需要显示的显露出来，而且GroupBy经常只需要一行代码就可以计算每组的和、均值、计数、最小值以及其他累计值。GroupBy的用处就是将这些步骤抽象：用户不需要知道底层如何进行计算，只要把操作成一个整体就够了

```python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns = ['key', 'data'])
df
```

```python
# 可以用DataFrame的groupby()方法进行绝大多数常见的分割-应用-组合操作，将需要分组的列名传进去即可
df.groupby('key')
```

df.groupby()返回对象不是一个DataFrame对象，而是一个DataFrameGroupBy对象，这个对象可以看成是一个特殊的DataFrame对象，里面隐藏着若干组数据，但是在没有应用累计函数之前不会计算。这种延迟计算（lazy evaluation）的方法使得大多数常见的累计操作可以通过一种对用户而言几乎是透明的方式非常高效地实现。

```python
# 对DataFrameGroupBy对象应用累计函数
df.groupby('key').sum()
```

#### 3. 累计、过滤、转换和应用

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
```

(1) 累计aggregate：支持字符串、函数或函数列表，并且能一次性计算 

```python
# 对key的值进行分组，计算每组里每一列的中位数
df.groupby('key').median()
```

```python
# 对key的值进行分组，一次性得到每组的每一列的最小值、中位数、最大值、总和
 # pandas累计函数都能用'min'这种方式调用，建议用此种方式，相比np.median这种方式更方便而且范围广，比如'count'就没有np.count，至于直接用min只有少数函数可以median就不行
df.groupby('key').aggregate(['min', np.median, max, np.sum])
```

```python
# 对key的值进行分组，一次性得到每组的data1列的最小值、中位数、最大值、总和
df.groupby('key')['data1'].aggregate(['min', np.median, max, np.sum])
```

另一种方法是通过Python字典指定不同列得到不同的累计函数

```python
# 对key的值进行分组，一次性得到每组里data1列的最小值和data2列的最大值
df.groupby('key').aggregate({'data1': min,
                             'data2': max})
```

```python
# 对key的值进行分组，一次性得到每组里data1列的最小值及个数和data2列的最大值及中位数
df.groupby('key').aggregate({'data1': ['min', 'count'],
                             'data2': ['max', 'median']})
```

(2) 过滤filter：可以按照分组的属性丢弃若干数据，分组后过滤有点类似SQL里的having

首先介绍基础的DataFrame.filter(items=None, like=None, regex=None, axis=None)

items：这是一个列表，包含你想要选择的列名。例如，df.filter(items=['A', 'B'])会返回一个新的DataFrame，只包含列名为'A'和'B'的列。

like：这是一个字符串，用于筛选列名包含特定字符串的列。例如，df.filter(like='A')会返回一个新的DataFrame，只包含列名中包含'A'的列。

regex：这是一个正则表达式，用于筛选列名匹配该正则表达式的列。例如，df.filter(regex='^A.*')会返回一个新的DataFrame，只包含列名以'A'开头的列。

axis：这是一个整数，用于指定应用筛选条件的轴。如果axis=0，则应用筛选条件到行标签；如果axis=1，则应用筛选条件到列标签。默认值为1。

注意：items、like和regex参数只能同时使用一个。

```python
print(df['data2'])
print(type(df['data2']))
print(type(df.filter(items = ['data2'])))
display('df', "df.filter(items = ['data2'])")
```

可以看到df['data2']取df的data2列返回结果是一个Series没有列名，如果没有这个列名会报错，而df.filter(items = ['data2'])返回结果仍是一个DataFrame，如果没有这个列名会返回空的DataFrame

现在来看Group后的filter用法的不同：groupby.filter(func, dropna=True, *args, **kwargs)

func：这是一个函数，它接受一个组作为输入，如果该组满足特定条件，就返回True，否则返回False。filter()函数会保留那些函数返回值为True的组。

dropna：这是一个布尔值，如果为True，则当func返回False时，filter()函数会删除所有的数据。如果为False，则当func返回False时，filter()函数会保留组的索引，但所有的列都会被填充为NaN。默认值为True。

*args, **kwargs：这些是传递给func的额外参数。

```python
# 自定义一个函数返回data2列标准差大于4的组
def filter_func(x):
    return x['data2'].std() > 4

# 将自定义函数返回的组输入filter中的fun函数中筛选过滤后返回原始df中的数据
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

groupby.filte()函数返回的是一个布尔值，表示每个组是否通过过滤。可以看到df中A组的data2列的标准差1.4不大于4，因此被过滤掉只返回B和C组

```python
df.groupby('key').aggregate(['mean', 'min'])
```

```python
# 筛选对key列分组后data1列的均值大于1而且data2列的最小值大于等于3的组
def filter_func(group):
    return group['data1'].mean() > 1 and group['data2'].min() >=3

df.groupby('key').filter(filter_func)
```

(3) 转换transform

累计返回的是对组内全量数据缩减过的结果，而转换操作会返回一个新的全量数据。数据经过转换后，其形状与原来的输入数据一样

常见的例子是将每一组的样本数据减去各组的均值，实现数据标准化

```python
df.groupby('key').transform(lambda x:x - x.mean())
```

(4) 应用apply：在每个组上应用任意方法。apply()输入一个DataFrame,返回一个Pandas对象(Series或DataFrame)或一个标量。组合操作会适应结果类型,其实就是返回任意自定义函数的结果。

```python
# 用apply()将第一列数据以第二列的和为基数进行标准化：x['data1'] = x['data1'] / x['data2'].sum(), data1列每个值除以data2列的和
def norm_by_data2(x):
    # x是一个分组数据的DataFrame
    x['data1'] /= x['data2'].sum()
    return x

display('df', "df.groupby('key').apply(norm_by_data2)")
```

## 10. 数据透视表

数据透视表将每一列数据作为输入，输出将数据不断细分成多个维度累计信息的二维数据表。分割组合发生在二维网格上（行列同时分组）

### 10.1 演示数据透视表

```python
# 在Seaborn程序库获取泰坦尼克号的乘客信息：性别gender, 年龄age, 船舱等级class, 船票价格fare paid 
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

### 10.2 手工制作数据透视表

```python
# 按照性别、最终生还状态或其他组合属性进行分组。例如用groupby实现统计不同性别乘客的生还率
titanic.groupby('sex')['survived'].mean()
```

可以看到女性生存率74%，而男性生存率只有18%；如果想进一步统计不同性别与船舱等级的生存率，用groupby实现如下：

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean')
```

```python
# 将最里面的行索引class转为列索引
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

由此可以看到一等舱和二等舱明显生存率更高，但这段代码有点复杂，由于二维的groupby应用场景非常普遍，因此Pandas提供了一个快捷方式pivot_table来快速解决多维累计分析任务

### 10.3 数据透视表语法

用DataFrame的pivot_table实现上面同样的效果：

```python
# 类似Excel的数据透视表：值是survived的均值，行标签(类别)index设为性别sex,列标签(系列)设为class
titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

与groupby相比，这行代码的可读性更强，取得结果一样。生还率最高的是一等舱的女性，几乎全部生还，而三等舱的男性生还率只有十分之一多一点。

1. 多级数据透视表

与groupby类似，数据透视表中的分组也可以通过各种参数指定多个等级，例如：通过pd.cut函数将年龄分段，再加入进入作为第三个维度。

```python
# 年龄分段(0, 18], (18, 80] 左开右闭
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

### 10.4 案例：美国人的生日

由美国疾病防治中心提供的公开生日数据，这些数据可以从https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv 下载，Andrew Gelman团队已经对这个数据集进行了深入分析，详情见http://andrewgelman.com/2012/06/14/cool-ass-signal-processing-using-gaussian-processes/

```python
births = pd.read_csv('births.csv')
births.head()
```

先增加一列表示不同年代，看看各年代的男女出生比列：

```python
births['decade'] = 10 * (births['year'] // 10)
births.pivot_table('births', index = 'decade', columns='gender', aggfunc = 'sum')
```

可以发现每个年代的男性出生率都比女性出生率高，将每一年出生人数画出来直观展现。

```python
# %matplotlib inline是一个在Jupyter Notebook或IPython环境中使用的魔术命令。它的作用是将matplotlib生成的图表直接嵌入到Notebook中的输出单元格中，而不是作为弹出窗口显示。
# 具体来说，当你调用matplotlib.pyplot的绘图函数（如plot()）进行绘图的时候，或者生成一个figure画布的时候，可以直接在你的python console里面生成图像。
# 这样，你就可以直接在Notebook中看到你的图像，而不需要调用plt.show()来显示图像
%matplotlib inline
import matplotlib.pyplot as plt
sns.set() # 使用Seaborn风格
births.pivot_table('births', index = 'year', columns = 'gender', aggfunc = 'sum').plot()
plt.ylabel('total births per year')
```

借助一个简单的数据透视表和plot()方法，马上就可以发现不同性别出生率的趋势，过去50年间男生出生率比女生出生率高5%。深入探索数据，对数据做一点清理工作，消除由于输错了日期而造成的异常点（如6月31号）或者是缺失值（如1999年6月）。消除这些异常的简便方法是直接剔除异常值。通过更稳定的sigma消除法（sigma-clippinig, 按照正太分布标准差划定范围，SciPy中默认是四个标准差）操作来实现：

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
print(quartiles)
mu = quartiles[1]
print(mu)
# 以下是样本均值的稳定性估计(robust estimate), 其中0.74是指标准正太分布的分位数间距，在query()方法中用这个范围可以将有效的生日数据筛选出来
sig = 0.74 * (quartiles[2] - quartiles[0])
print(sig)
```

```python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births
```

```python
# 然后将day列设置为整数，这列数据在筛选之前是字符串，因为数据集中有的列含有缺失值'null'
births['day'] = births['day'].astype(int)
```

现在就可以将年月日组合起来创建一个日期索引了，这样就可以快速计算每一行是星期几：

```python
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format = '%Y%m%d')
births['dayofweek'] = births.index.dayofweek
births
```

```python
# 用这个索引可以画出不同年代不同星期的日均出生数据
import matplotlib.pyplot as plt
import matplotlib as mpt

births.pivot_table('births', index = 'dayofweek', columns = 'decade', aggfunc = 'mean').plot()
# plt.gca().set_xticklabels(xticks = range(7), xticklabels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
# plt.ylabel('mean births by day')
plt.gca().set(xticks = range(7),
              xticklabels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day');
```

由图可知，周末的出生人数比工作日要低很多。另外CDC只提供了1989年之前的数据，所以没有20世纪90年代和21世纪的数据。

另一个有趣的图表是画出各个年份平均每天的出生人数，可以按照月和日两个维度分别对数据进行分组：

```python
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.head()
```

这是一个包含月和日的多级索引。为了让数据用图形表示，我们可以虚构一个年份，与月和日组成新索引（注意日期为2月29日时，索引年份需要用闰年，例如2012年）：

```python
from datetime import datetime

births_by_date.index = [datetime(2012, month, day)
                        for (month, day) in births_by_date.index]
births_by_date.head()
```

```python
# 用plot()方法将结果画成图
fig, ax = plt.subplots(figsize = (12, 4))
births_by_date.plot(ax = ax)
```

从图中可以明显看到，在美国节假日的时候，出生人数急速下降（例如美国独立日7月4号，劳动节，感恩节，圣诞节和新年）。这种情况可能是由于医院放假导致的接生减少，而非某种自然生育的心理学效应。关于这个趋势更详细的介绍可参考Andrew Gelman的博客http://andrewgelman.com/2012/06/14/cool-ass-signal-processing-using-gaussian-processes/
