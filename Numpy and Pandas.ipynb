## 5. 广播

广播可以简单理解为用于不同大小数组的二元通用函数（加、减、乘等）的一组规则，同之前的常用通用函数一样，都是通过向量化的操作减少缓慢的python循环

### 5.1 广播介绍

```python
# 对于同样大小的数组，二元运算符是对相应元素逐个计算
a = np.array([0, 1, 2])
b = np.array([5, 5, 5])
print(a + b)
# 广播允许二元运算符用于不同大小的数组，例如简单将一个标量（可以理解为一个零维的数组）和一个数组相加
# 可以理解为先将数值5扩展为数组[5, 5, 5]然后执行加法
print(a + 5)
```

```python
# 将一维数组array([0, 1, 2])扩展成匹配3x3的二维数组array([[0, 1, 2],[0, 1, 2],[0, 1, 2]])
M = np.ones((3, 3))
print(M)
print(a)
M + a
```

```python
# 两个数组同时广播
a = np.arange(3)
b = np.arange(3)[:, np.newaxis]
print(a)
print(b)
a + b
```

### 5.2 广播规则

规则1：如果两个数组的维度数不同，那么小维度数组的形状将会在最左边补1

规则2：如果两个数组的形状在任何一个维度上都不匹配，那么数组的维度会沿着维度为1的维度扩展以匹配另外一个数组的形状

规则3：如果两个数组的形状在任何一个维度上都不匹配并且没有任何一个维度等于1，那么就会引发异常

```python
# 广播示例 1 ：将一个二维数组和一个一维数组相加
M = np.ones((2, 3))
a = np.arange(3)
print(M, a)
print(M.shape, a.shape)
print('根据规则1，数组a的维度更小，所以在其左边补1，m.shape(2, 3) a.shape(1, 3)')
print('根据规则2，两个数组第一个维度不匹配，扩展数组a以匹配数组，a.shape(2, 3)')
M + a
```

```python
# 广播示例 2：两个数组均需要广播
a = np.arange(3).reshape(3, 1)
b = np.arange(3)
print(a, b)
print(a.shape, b.shape)
print('根据规则1，数组b的维度更小，所以在其左边补1，a.shape(3, 1) b.shape(1, 3)')
print('根据规则2，两个数组维度不匹配，扩展两个数组，a.shape(3, 3) b.shape(3, 3)')
print('''
[[0, 0, 0],
 [1, 1, 1],
 [2, 2, 2]] + [[0, 1, 2],
               [0, 1, 2],
               [0, 1, 2]] = \n''', a + b)
```

### 5.3 广播实际应用

1. 数组的归一化

数组归一化是将数组的数值范围缩放到特定范围或标准化的过程。在数据处理和机器学习中，常常需要对数据进行归一化，以便使数据具有一致的尺度和范围，从而有助于模型的训练和预测。

常见的数组归一化方法包括：

1. 最小-最大归一化（Min-Max Scaling）： 这是最常见的归一化方法之一。它将数据缩放到指定的范围，通常是 [0, 1]。

2. 标准化（Standardization）： 这种方法将数据转换为均值为 0、标准差为 1 的正态分布（也叫标准正态分布）。标准化使得数据分布更符合统计假设，有助于某些机器学习算法的性能。

为什么要进行数组归一化？

1. 消除尺度影响： 不同的特征可能具有不同的尺度和范围，这可能会对某些机器学习算法产生负面影响。通过归一化，可以消除特征之间的尺度差异，使得算法更加稳定。

2. 加速收敛： 对于许多优化算法（如梯度下降），归一化可以加速算法的收敛过程，使其更快地达到最优解。

3. 提高模型性能： 在某些情况下，归一化可以提高机器学习模型的性能，因为模型更容易学习和泛化归一化后的数据。

4. 可视化和解释性： 归一化后的数据更容易可视化和解释，因为它们具有相似的尺度。

需要注意的是，在某些情况下，不需要进行数组归一化，特别是当数据已经在合适的尺度范围内时，或者某些特定的机器学习算法不敏感于尺度差异时。因此，在进行数组归一化之前，需要仔细考虑数据的性质和算法的要求。

rand.multivariate_normal: 这个函数生成多元正态分布的随机数，也被称为多维高斯分布。生成的随机数是多维的，通常用于生成多维数据，如多维特征的数据集。接受均值向量和协方差矩阵作为参数，用于定义多维正态分布的性质。成的随机数是一个二维数组，其中每一行代表一个多维随机数据点。

np.random.normal: 这个函数生成单变量正态分布（univariate normal distribution）的随机数，也被称为高斯分布。生成的随机数是单个值，通常用于生成一维数据或单个随机变量的值。接受均值和标准差作为参数，用于定义单变量正态分布的性质。生成的随机数是一个一维数组，包含单变量随机数值。

### 7.5 示例：数据区间划分，手动创建直方图

```python
# 用ufun.c.at,假定有1000个值，快速统计分布在各个区间中的数据频次
np.random.seed(42)
x = np.random.randn(100)

# 手动计算直方图，生成一个[-5, 5]之间均匀生成的20个数据点，划分成20个均匀的小区间
bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins) # 生成一个和bins数组形状一样的全零数组
print('bins:', bins)
print('counts:', counts)

# 为每个x找到合适的区间
"""
np.searchsorted(bins, x): 这是调用NumPy中的np.searchsorted函数，该函数接受两个参数：
第一个参数 bins 是已排序的数组，表示被查找的范围。
第二个参数 x 是要查找的数值。
np.searchsorted函数返回的结果是一个整数，表示在将 x 插入到 bins 数组中时，应该插入的位置。具体来说：
如果 x 小于 bins 中的所有值，那么返回值将是0，表示 x 应该插入到 bins 数组的开头。
如果 x 大于 bins 中的所有值，那么返回值将是 len(bins)，表示 x 应该插入到 bins 数组的末尾。
如果 x 介于 bins 中的两个值之间，那么返回值将是介于它们之间的位置，以保持 bins 的升序顺序。
"""
i = np.searchsorted(bins, x)

# 为每个区间加上1, 计数数组counts反映的是在各个区间中的点的个数
np.add.at(counts, i, 1)

# 画出结果
plt.plot(bins, counts, linestyle = '-')
plt.show()

# 用plt.hist()一步实现上述效果
plt.hist(x, bins, histtype = 'step')
```

## 8. 数组的排序

```python
# 简单的选择排序重复寻找列表中的最小值，并且不断交换直到列表是有序的
def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
        return x
    
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
```

### 8.3 示例：K个最近邻

以下示例展示如何使用argsort函数沿着多个轴快速找到集合中每个点的最近邻。

```python
# 首先在二维平面上创建一个有10个随机点的集合，按照惯例这些数据点放在一个10X2的数组中
X = rand.rand(10, 2)
print('X: \n', X)

# 将这些点画散点图
import matplotlib.pyplot as plt
import seaborn; seaborn.set # 设置绘图风格

plt.scatter(X[:, 0], X[:, 1], s = 100)
```

现在来计算两两数据点对间的距离，两点间的距离的平方等于每个维度距离差的平方的和，利用广播和聚合功能计算矩阵的平方距离。

```python
dist_sq = np.sum((X[:,np.newaxis,:] - X[np.newaxis,:,:]) **2, axis = -1)
dist_sq.shape
```

```python
# 确认以上步骤，应该看到该矩阵的对角线（也就是每个点到其自身的距离）都是0
dist_sq.diagonal()
```

```python
# 得到一个转化为两点间的平方距离的矩阵后，就可以使用np.argsort函数沿着每行进行排序，最左边的列给出的索引值就是最近邻
# 需要注意的是，第一列是按照0-9从小到大顺序排列的，这是因为每个点的最近邻是其自身
nearest = np.argsort(dist_sq, axis = 1)
nearest
```

如果使用全排序，我们实际上可以实现的比这个例子展示的更多。如果仅仅关心k个最近邻，那么唯一需要做的是分隔每一行，这样最小的k + 1的平方距离将排在最前面，其他更长距离占据矩阵该行的其他位置。可以用np.argpartition函数实现：

```python
k = 2
nearest_partition = np.argpartition(dist_sq, k + 1, axis = 1)
nearest_partition
```

```python
# 为了将邻节点网络可视化，我们将每个点和其最近的两个最近邻连接
plt.scatter(X[:, 0], X[:, 1], s = 100)

k = 2
for i in range(X.shape[0]):
    for j in nearest_partition[i, :k+1]:
        # 画一条从X[i]到X[j]的线段
        # 用zip方法实现
        plt.plot(*zip(X[j], X[i]), color = 'black')
```

一种方法是通过一个覆盖全局的掩码表示缺失值，另一种方法是用一个标签值表示缺失值

### 5.2 Pandas的缺失值

Pandas选择用标签方法表示缺失值，包括两种Python原有的缺失值：浮点数据类型的NaN值，以及Python的None对象

```python
# 1.None：Python对象类型的缺失值，只能用于'object'数组类型，即由python对象构成的数组
vals1 = np.array([1, None, 3, 4])
vals1
```

```python
for dtype in ['object', 'int']:
    print('dtype = ', dtype)
    %timeit np.arange(1E6, dtype = dtype).sum()
    print()
```

这里dtype = object表示Numpy认为这个数组是Python对象构成的，因此其类型判断为object,这种类型比其他原生类型数组要消耗更多资源。

Python中没有定义整数与None之间的加法预算，因此对使用Python对象构成的包含None的数组进行累计操作sum()或min()等会报错

```python
# 2.NaN:数值类型的缺失值，全称Not a Number,在任何系统都兼容的特殊浮点数
vals2 = np.array([1, np.nan, 3, 4])
print(vals2.dtype)
# 无论和NaN进行任何操作，最终结果都是NaN
print('1 + np.nan :', 1 + np.nan)
print('0 * np.nan :', 0 * np.nan)
print(vals2.sum(), vals2.min(), vals2.max())
# 使用Numpy特殊的累计函数可以忽略缺失值的影响
print(np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2))
```

```python
# Pandas中NaN与None的差异：Pandas看作等价交换，会将没有标签的数据类型,或None自动转换为NaN
print(pd.Series([1, np.nan, 2, None]))
x = pd.Series(range(2), dtype = int)
print('x: \n', x)
x[0] = None
x
```

Pandas对不同类型缺失值的转换规则:

|类型                 | 缺失值转换规则               | NA 标签值                     |
|--------------       |-----------------------------|------------------------|

| ``object对象类型``  | 无变化                   | ``None`` or ``np.nan`` |
| ``integer整数类型`` | 强制转换为 ``float64``         | ``np.nan``             |

## 6. 层级索引

Series和DataFrame主要是对一维和二维数据储存，对于多维数据，Pandas提供了Panel和Panel4D对象解决三维数据和四维数据；但在实践中，更直观的形式是通过层级索引（多级索引）配合多个有不同等级的一级索引一起使用，将多维数组转化为类似一维Series和二维DataFrame的形式

### 6.1 多级索引Series

如何用一维的Series对象表示二维数据：用一系列包含特征与取值的数据点进行简单演示

```python
# 1.笨方法
# 假设想要分析美国各州在不同年份的数据，可以用python元组来表示索引
index = [('加州', 2000), ('加州', 2010), ('纽约', 2000), ('纽约', 2010), ('新泽西', 2000), ('新泽西', 2010)]
populations = [33871648, 37253956, 18986457, 19378102, 20851820, 25145561]
pop = pd.Series(populations, index = index)
print(pop)
# 通过元组构成的多级索引，直接在Series上取值和用切片查询数据
print('取加州2000至新泽西2000的数据')
print(pop[('加州', 2000):('新泽西', 2000)])
# 这样做很不方便，假如要取所有2010年的数据，需要用循环找出2010年
for i in pop.index:
    if i[1] == 2010:
        print(i)

pop[[i for i in pop.index if i[1] == 2010]] # 方法不够简洁，处理较大数据时不够高效
```

```python
# 好方法：Pandas多级索引，MultiIndex类型提供了更丰富的方法，用元组创建一个多级索引
index = pd.MultiIndex.from_tuples(index)
# levels是一个列表，元素为两个子列表，一个列表包含州名，一个列表包含年份
print(index)
print('将pop的索引重置为MultiIndex，就会看到层级索引')
pop  = pop.reindex(index)
print(pop) 
# 前两列表示Series的多级索引值，第三列是数据，可以直接用第二个索引获得2010年的数据，与Pandas的切片查询用法一致
pop[:, 2010] # 结果是一个单索引的数组，与笨方法相比语法更简单操作更方便
```

```python
# 高维数据的多级索引,unstack()快速将一个多级索引Series转化为普通索引的DataFrame, stack()相反
print(pop)
print('将一个多级索引的Series转化为普通索引的DataFrame')
print(pop.unstack())
print('将一个普通索引的DataFrame转化为多级索引的Series')
print(pop.unstack().stack())
# 能用一个多级索引的Series表示二维数据能用一个多级索引的Series或者DataFrame表示三维或更高维数据
pop_df = pd.DataFrame({'total':pop, 'under18': [9267089, 9284094, 4687374, 4318033, 5906301, 6879014]})
pop_df
```

```python
# Pandas数据运算方法介绍过的通用函数和其他功能同样适用于层级索引，例如计算18岁以下的人口占总人口的比例
f_u18 = pop_df['under18'] / pop_df['total']
f_u18.unstack()
```

```python
# 定义一个能够创建DataFrame某种形式的函数
def make_df(cols, ind):
    """一个简单的DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

# DataFrame示例
make_df('ABC', range(3))
```

### 7.2 通过pd.concat实现简易合并

```python
# pd.concat(
#     objs: 'Iterable[NDFrame] | Mapping[HashableT, NDFrame]',
#     *,
#     axis: 'Axis' = 0,
#     join: 'str' = 'outer',
#     ignore_index: 'bool' = False, 设置True会忽略索引合并后创建一个新的整数索引
#     keys=None, 设置多级索引标签
#     levels=None,
#     names=None,
#     verify_integrity: 'bool' = False, 设置True时有重复索引会报错
#     sort: 'bool' = False,
#     copy: 'bool | None' = None,
# ) -> 'DataFrame | Series'

# pd.concat()可以简单的合并一维的Series或DataFrame对象，与np.concatenate()合并数组一样
ser1 = pd.Series(['A', 'B', 'C'], index = [1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index = [4, 5, 6])
print(pd.concat([ser1, ser2]))

# 合并高维数据
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])
print(df1)
print(df2)
print(pd.concat([df1, df2]))

# pd.concat默认axis = 0即沿着行方向合并，也可以设置沿着列方向拼接
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
print(df3)
print(df4)
print(pd.concat([df3, df4]))
pd.concat([df3, df4], axis = 1)
```

### 8.2 数据连接的类型

pd.merge()函数实现了三种数据连接的类型：一对一、多对一和多对多

```python
import pandas as pd
import numpy as np

# 设置展现display输出并排显示结果
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
# 1.一对一连接，按列合并，类似数据库的innerjoin自动以相同的列作为键进行连接, 默认丢弃原来的行索引
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
df3 = pd.merge(df1, df2)
display('df1', 'df2', 'df3')
```

### 9.3 GroupBy:分割、应用和组合

对某些标签或索引的局部进行累计分析需要用到groupby, 虽然分组(group by)这个名字借用SQL，其理念引用发明R语言frame的Hadley Wickkam的观点更适合：

分割（split）、应用（apply）、组合（combine）

#### 1. 分割、应用和组合

下图清晰描述了GroupBy的过程

-分割：将DataFrame按照指定的键key分割成若干组

-应用：每个组应用函数，通常是累计、转换或过滤函数

-组合：将每一组的结果合并成一个输出数组

中间分割过程不需要显示的显露出来，而且GroupBy经常只需要一行代码就可以计算每组的和、均值、计数、最小值以及其他累计值。GroupBy的用处就是将这些步骤抽象：用户不需要知道底层如何进行计算，只要把操作成一个整体就够了

```python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns = ['key', 'data'])
df
```

```python
# 可以用DataFrame的groupby()方法进行绝大多数常见的分割-应用-组合操作，将需要分组的列名传进去即可
df.groupby('key')
```

df.groupby()返回对象不是一个DataFrame对象，而是一个DataFrameGroupBy对象，这个对象可以看成是一个特殊的DataFrame对象，里面隐藏着若干组数据，但是在没有应用累计函数之前不会计算。这种延迟计算（lazy evaluation）的方法使得大多数常见的累计操作可以通过一种对用户而言几乎是透明的方式非常高效地实现。

```python
# 对DataFrameGroupBy对象应用累计函数
df.groupby('key').sum()
```

#### 3. 累计、过滤、转换和应用

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
```

(1) 累计aggregate：支持字符串、函数或函数列表，并且能一次性计算 

```python
# 对key的值进行分组，计算每组里每一列的中位数
df.groupby('key').median()
```

```python
# 对key的值进行分组，一次性得到每组的每一列的最小值、中位数、最大值、总和
 # pandas累计函数都能用'min'这种方式调用，建议用此种方式，相比np.median这种方式更方便而且范围广，比如'count'就没有np.count，至于直接用min只有少数函数可以median就不行
df.groupby('key').aggregate(['min', np.median, max, np.sum])
```

```python
# 对key的值进行分组，一次性得到每组的data1列的最小值、中位数、最大值、总和
df.groupby('key')['data1'].aggregate(['min', np.median, max, np.sum])
```

另一种方法是通过Python字典指定不同列得到不同的累计函数

```python
# 对key的值进行分组，一次性得到每组里data1列的最小值和data2列的最大值
df.groupby('key').aggregate({'data1': min,
                             'data2': max})
```

```python
# 对key的值进行分组，一次性得到每组里data1列的最小值及个数和data2列的最大值及中位数
df.groupby('key').aggregate({'data1': ['min', 'count'],
                             'data2': ['max', 'median']})
```

(2) 过滤filter：可以按照分组的属性丢弃若干数据，分组后过滤有点类似SQL里的having

首先介绍基础的DataFrame.filter(items=None, like=None, regex=None, axis=None)

items：这是一个列表，包含你想要选择的列名。例如，df.filter(items=['A', 'B'])会返回一个新的DataFrame，只包含列名为'A'和'B'的列。

like：这是一个字符串，用于筛选列名包含特定字符串的列。例如，df.filter(like='A')会返回一个新的DataFrame，只包含列名中包含'A'的列。

regex：这是一个正则表达式，用于筛选列名匹配该正则表达式的列。例如，df.filter(regex='^A.*')会返回一个新的DataFrame，只包含列名以'A'开头的列。

axis：这是一个整数，用于指定应用筛选条件的轴。如果axis=0，则应用筛选条件到行标签；如果axis=1，则应用筛选条件到列标签。默认值为1。

注意：items、like和regex参数只能同时使用一个。

```python
print(df['data2'])
print(type(df['data2']))
print(type(df.filter(items = ['data2'])))
display('df', "df.filter(items = ['data2'])")
```

可以看到df['data2']取df的data2列返回结果是一个Series没有列名，如果没有这个列名会报错，而df.filter(items = ['data2'])返回结果仍是一个DataFrame，如果没有这个列名会返回空的DataFrame

现在来看Group后的filter用法的不同：groupby.filter(func, dropna=True, *args, **kwargs)

func：这是一个函数，它接受一个组作为输入，如果该组满足特定条件，就返回True，否则返回False。filter()函数会保留那些函数返回值为True的组。

dropna：这是一个布尔值，如果为True，则当func返回False时，filter()函数会删除所有的数据。如果为False，则当func返回False时，filter()函数会保留组的索引，但所有的列都会被填充为NaN。默认值为True。

*args, **kwargs：这些是传递给func的额外参数。

```python
# 自定义一个函数返回data2列标准差大于4的组
def filter_func(x):
    return x['data2'].std() > 4

# 将自定义函数返回的组输入filter中的fun函数中筛选过滤后返回原始df中的数据
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

groupby.filte()函数返回的是一个布尔值，表示每个组是否通过过滤。可以看到df中A组的data2列的标准差1.4不大于4，因此被过滤掉只返回B和C组

```python
df.groupby('key').aggregate(['mean', 'min'])
```

```python
# 筛选对key列分组后data1列的均值大于1而且data2列的最小值大于等于3的组
def filter_func(group):
    return group['data1'].mean() > 1 and group['data2'].min() >=3

df.groupby('key').filter(filter_func)
```

(3) 转换transform

累计返回的是对组内全量数据缩减过的结果，而转换操作会返回一个新的全量数据。数据经过转换后，其形状与原来的输入数据一样

常见的例子是将每一组的样本数据减去各组的均值，实现数据标准化

```python
df.groupby('key').transform(lambda x:x - x.mean())
```

(4) 应用apply：在每个组上应用任意方法。apply()输入一个DataFrame,返回一个Pandas对象(Series或DataFrame)或一个标量。组合操作会适应结果类型,其实就是返回任意自定义函数的结果。

```python
# 用apply()将第一列数据以第二列的和为基数进行标准化：x['data1'] = x['data1'] / x['data2'].sum(), data1列每个值除以data2列的和
def norm_by_data2(x):
    # x是一个分组数据的DataFrame
    x['data1'] /= x['data2'].sum()
    return x

display('df', "df.groupby('key').apply(norm_by_data2)")
```

## 10. 数据透视表

数据透视表将每一列数据作为输入，输出将数据不断细分成多个维度累计信息的二维数据表。分割组合发生在二维网格上（行列同时分组）

### 10.1 演示数据透视表

```python
# 在Seaborn程序库获取泰坦尼克号的乘客信息：性别gender, 年龄age, 船舱等级class, 船票价格fare paid 
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

### 10.2 手工制作数据透视表

```python
# 按照性别、最终生还状态或其他组合属性进行分组。例如用groupby实现统计不同性别乘客的生还率
titanic.groupby('sex')['survived'].mean()
```

可以看到女性生存率74%，而男性生存率只有18%；如果想进一步统计不同性别与船舱等级的生存率，用groupby实现如下：

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean')
```

```python
# 将最里面的行索引class转为列索引
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

由此可以看到一等舱和二等舱明显生存率更高，但这段代码有点复杂，由于二维的groupby应用场景非常普遍，因此Pandas提供了一个快捷方式pivot_table来快速解决多维累计分析任务

### 10.3 数据透视表语法

用DataFrame的pivot_table实现上面同样的效果：

```python
# 类似Excel的数据透视表：值是survived的均值，行标签(类别)index设为性别sex,列标签(系列)设为class
titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

与groupby相比，这行代码的可读性更强，取得结果一样。生还率最高的是一等舱的女性，几乎全部生还，而三等舱的男性生还率只有十分之一多一点。

1. 多级数据透视表

与groupby类似，数据透视表中的分组也可以通过各种参数指定多个等级，例如：通过pd.cut函数将年龄分段，再加入进入作为第三个维度。

```python
# 年龄分段(0, 18], (18, 80] 左开右闭
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

借助一个简单的数据透视表和plot()方法，马上就可以发现不同性别出生率的趋势，过去50年间男生出生率比女生出生率高5%。深入探索数据，对数据做一点清理工作，消除由于输错了日期而造成的异常点（如6月31号）或者是缺失值（如1999年6月）。消除这些异常的简便方法是直接剔除异常值。通过更稳定的sigma消除法（sigma-clippinig, 按照正太分布标准差划定范围，SciPy中默认是四个标准差）操作来实现：

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
print(quartiles)
mu = quartiles[1]
print(mu)
# 以下是样本均值的稳定性估计(robust estimate), 其中0.74是指标准正太分布的分位数间距，在query()方法中用这个范围可以将有效的生日数据筛选出来
sig = 0.74 * (quartiles[2] - quartiles[0])
print(sig)
```

```python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births
```

```python
# 然后将day列设置为整数，这列数据在筛选之前是字符串，因为数据集中有的列含有缺失值'null'
births['day'] = births['day'].astype(int)
```

这是一个包含月和日的多级索引。为了让数据用图形表示，我们可以虚构一个年份，与月和日组成新索引（注意日期为2月29日时，索引年份需要用闰年，例如2012年）：

## 11. 向量化字符串操作

向量化字符串是指在处理字符串数据时，使用向量化操作来提高性能和效率。在传统的字符串处理中，通常需要使用循环或迭代来处理字符串的每个元素，这可能会导致代码执行缓慢，特别是当处理大量字符串数据时。

向量化字符串操作使用底层的数学库来执行相同的操作，但是它是在整个字符串数组上进行的，而不是在单个字符串上循环。这允许利用底层的优化，提高了处理速度，尤其是对于大规模数据。

### 11.1 Pandas字符串操作简介

```python
import numpy as np
x = np.array([2, 3, 5, 7, 11, 13])
x * 2 
```

向量化操作简化了纯数值的数组操作语法——我们需要再担心数组的长度或维度，只需要关心需要的操作。然而，由于Numpy并没有为字符串数组提供简单的接口，因此需要用for循环来解决问题：

```python
data = ['peter', 'Paul', 'MARY', 'gUIDO']
# 将字符串的首字母大写
[s.capitalize() for s in data]
```

这种做法对某些数据可能是有效的，但是假如数据中出现了缺失值data = data = ['peter', 'Paul', None, 'MARY', 'gUIDO']，那么这样做就会报错。

Pandas为包含字符串的Series和Index对象提供了str属性堪称两全其美的方法，既可以满足向量化字符串操作的需求，又可以正确处理缺失值。

```python
import pandas as pd
data = ['peter', 'Paul', None, 'MARY', 'gUIDO']
names = pd.Series(data)
names
```

```python
# 直接调用转换大写方法captitalize()将所有字符串首字母大写, 缺失值会被跳过
names.str.capitalize()
```

```python
# str.contains检查是否包含字符串'peter'，包含返回True
names.str.contains('peter')
```

## 12. 处理时间序列

由于 Pandas 最初是为金融模型而创建的，因此它拥有一些功能非常强大的日期、时间、带时间索引数据的处理工具。本节将介绍的日期与时间数据主要包含三类。

- 时间戳：表示某个具体的时间点（例如 2015 年 7 月 4 日上午 7点）。

- 时间间隔与周期：表示开始时间点与结束时间点之间的时间长度，例如 2015 年（指的是 2015 年 1 月 1 日至 2015 年 12 月 31 日这段时间间隔）。周期通常是指一种特殊形式的时间间隔，每个间隔长度相同，彼此之间不会重叠（例如，以 24 小时为周期构成每一天）。

- 时间增量（time delta）或持续时间（duration）：表示精确的时间长度（例如，某程序运行持续时间 22.56 秒）。

在本节内容中，我们将介绍 Pandas 中的 3 种日期 / 时间数据类型的具体用法。由于篇幅有限，后文无法对 Python 或 Pandas 的时间序列工具进行详细的介绍，仅仅是通过一个宽泛的综述，总结何时应该使用它们。在开始介绍 Pandas 的时间序列工具之前，我们先简单介绍一下 Python处理日期与时间数据的工具。在介绍完一些值得深入学习的资源之后，再通过一些简短的示例来演示 Pandas 处理时间序列数据的方法。

### 12.1 Python的日期与时间工具

在 Python 标准库与第三方库中有许多可以表示日期、时间、时间增量和时间跨度（timespan）的工具。尽管 Pandas 提供的时间序列工具更适合用来处理数据科学问题，但是了解 Pandas 与 Python 标准库以及第三方库中的其他时间序列工具之间的关联性将大有裨益。

1. 原生Python的日期和时间工具：datatime 与 dateutil

Python 基本的日期与时间功能都在标准库的 datetime 模块中。如果和第三方库 dateutil 模块搭配使用，可以快速实现许多处理日期与时间的功能。例如可以用 datetime 类型创建一个日期：

有了datetime对象就可以进行很多操作，例如打印出这一天是星期几：

```python
# strftime是string format time的缩写，它接受一个格式化字符串作为参数，并返回一个格式化后的字符串表示日期时间。在这里，'%A' 是格式化字符串中的一部分，它表示星期几的完整名称
date.strftime('%A')
```

```python
# 这里为了打印出是星期几用了一个标准字符串格式代码'%A'，类似的还有
# 日期和时间的组合：
# %Y：四位数的年份（如 2022）。
# %y：两位数的年份（如 22）。
# %m：两位数的月份（01 到 12）。
# %B：完整的月份名称（如 January）。
# %b：缩写的月份名称（如 Jan）。
# %d：两位数的日期（01 到 31）。

# 时钟时间：
# %H：24小时制的小时数（00 到 23）。
# %I：12小时制的小时数（01 到 12）。
# %M：分钟数（00 到 59）。
# %S：秒数（00 到 59）。

# 时区和其他信息：
# %Z：时区的名称。
# %j：一年中的第几天。
# %U：一年中的第几周（以周日为一周的开始）。
# %W：一年中的第几周（以周一为一周的开始）。

# 周几：
# %A：星期几的完整名称。
# %a：星期几的缩写名称。
```

datetime 和 dateutil 模块在灵活性与易用性方面都表现出色，你可以用这些对象及其相应的方法轻松完成你感兴趣的任意操作。但如果你处理的时间数据量比较大，那么速度就会比较慢。就像之前介绍过的 Python 的原生列表对象没有 NumPy 中已经被编码的数值类型数组的性能好一样，Python 的原生日期对象同样也没有NumPy 中已经被编码的日期（encoded dates）类型数组的性能好。

2. 时间类型数组：Numpy的datetime64类型

Python 原生日期格式的性能弱点促使 NumPy 团队为 NumPy 增加了自己的时间序列类型。 datetime64 类型将日期编码为 64 位整数，这样可以让日期数组非常紧凑（节省内存）。 datetime64 需要在设置日期时确定具体的输入类型：

```python
import numpy as np
date = np.array('2015-07-04', dtype = np.datetime64)
date
```

有了这个日期格式，就可以进行快速的向量化操作：

```python
print(np.arange(12))
date + np.arange(12)
```

因为 NumPy 的 datetime64 数组内元素的类型是统一的，所以这种数组的运算速度会比 Python 的 datetime 对象的运算速度快很多，尤其是在处理较大数组时。

datetime64 与 timedelta64 对象的一个共同特点是，它们都是在基本时间单位（fundamental time unit）的基础上建立的。由于datetime64 对象是 64 位精度，所以可编码的时间范围可以是基本单元的 2的64次方倍。也就是说，datetime64 在时间精度（timeresolution）与最大时间跨度（maximum time span）之间达成了一种平衡。

比如你想要一个时间纳秒（nanosecond，ns）级的时间精度，那么你就可以将时间编码到 0~264 纳秒或 600 年之内，NumPy 会自动判断输入时间需要使用的时间单位。例如:

```python
# 以天为单位的日期
print(np.datetime64('2015-07-04'))
# 以分钟为单位的日期
print(np.datetime64('2015-07-04 12:00'))
```

需要注意的是，时区将自动设置为执行代码的操作系统的当地时区。你可以通过各种格式的代码设置基本时间单位。例如，将时间单位设置为纳秒：

```python
np.datetime64('2015-07-04 12:59:59.50', 'ns')
```

1. 重新取样与频率转换

处理时间序列数据时，经常需要按照新的频率（更高频率、更低频率）对数据进行重新取样。你可以通过 resample() 方法解决这个问题，或者用更简单的 asfreq() 方法。这两个方法的主要差异在于resample() 方法是以数据累计（data aggregation）为基础，而 asfreq() 方法是以数据选择（data selection）为基础。

看到 Google 的收盘价之后，让我们用两种方法对数据进行向后取样（down-sample）。这里用年末（'BA'，最后一个工作日）对数据进行重新取样（如图 3-6 所示）：

```python
# 比较原始数据、年度均值和年度重新取样的数据，通过不同的线型和颜色区分它们，并在图表上添加了图例。
# 绘制原始数据，透明度为0.5，线型为实线
goog.plot(alpha = 0.5, style = '-')

# 对数据进行年度重新采样，并计算每年的均值，线型为点线
goog.resample('BA').mean().plot(style = ':')

# 对数据按照年度频率进行重新取样，线型为虚线
goog.asfreq('BA').plot(style = '--')

# 添加图例，指定标签和位置左上角
plt.legend(['iput', 'resample', 'asfreq'], loc = 'upper left');
```

这两种取样方法的差异：在每个数据点上，resample 反映的是上一年的均值，而 asfreq 反映的是上一年最后一个工作日的收盘价。

在进行向前取样（up-sampling）时，resample() 与 asfreq() 的用法大体相同，不过重新取样有许多种配置方式。操作时，两种方法都默认将向前取样作为缺失值处理，也就是说在里面填充 NaN。与前面介绍过的 pd.fillna() 函数类似，asfreq() 有一个method 参数可以设置填充缺失值的方式。下面将对工作日数据按天进行重新取样（即包含周末）

```python
fig, ax = plt.subplots(2, sharex=True)
data = goog.iloc[:10]

data.asfreq('D').plot(ax=ax[0], marker='o')

data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(["back-fill", "forward-fill"],  fontsize = 'small')
plt.title('asfreq()向前填充与向后填充缺失值的结果对比')
```

上面那幅图是原始数据：非工作日的股价是缺失值，所以不会出现在图上。而下面那幅图通过向前填充与向后填充这两种方法填补了缺失值。

2. 时间迁移

另一种常用的时间序列操作是对数据按时间进行迁移。Pandas 原有两种解决这类问题的方法：shift() 和 tshift()。简单来说，shift() 就是迁移数据，而 tshift() 就是迁移索引。两种方法都是按照频率代码进行迁移。版本更新后tshit已被弃用，可以用shit也能实现迁移索引

```python
# shirt(periods,指定要移动的步数，可以是正数表示向前移动，负数表示向后移动
#       fre1,用于指定时间频率，主要在处理时间序列数据时使用。默认为 None，表示不考虑时间频率。
#       axis,用于指定移动的方向，可选值为 0 或 1。默认为 0，表示沿着索引的方向移动。
#       fill_value,用于指定缺失值的填充值，默认为 NaN。
#       )
```

下面用 shift() 让数据迁移900天，以及迁移索引900天：

```python
# 用shift()让数据迁移900天，以及迁移索引900天

# 创建一个包含2个子图的图表
fig, ax = plt.subplots(3, sharey=True)

# 对数据进行日频率重新采样，用前一天的值填充缺失值
goog = goog.asfreq('D', method='pad')

# 绘制原始数据
goog.plot(ax=ax[0])
# 将数据向后平移900天并绘制
goog.shift(900).plot(ax=ax[1])
# 使用shift函数替换tshift函数
goog.shift(900, freq='D').plot(ax=ax[2])

# 图例和注释
local_max = pd.to_datetime('2007-11-05') # 本地最大值的日期
offset = pd.Timedelta(900, 'D') # 时间偏移量

# 子图1
ax[0].legend(['input'], loc=2) # 图例在左上角
ax[0].get_xticklabels()[2].set(weight='heavy', color='red') # 第三个X轴加粗红色
ax[0].axvline(local_max, alpha=0.3, color='red') # 在本地最大值处画垂直线，设置透明度和颜色
# 子图2
ax[1].legend(['shift(900)'], loc=2)
ax[1].get_xticklabels()[2].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')
# 子图3
ax[2].legend(['shift(900)'], loc=2) 
ax[2].get_xticklabels()[1].set(weight='heavy', color='red')
ax[2].axvline(local_max + offset, alpha=0.3, color='red')

# 调整子图之间的间距
plt.subplots_adjust(hspace=0.3)
```

shift(900) 将数据向前推进了 900 天，这样图形中的一段就消失了（最左侧就变成了缺失值），而 shift(900, freq = 'D') 方法是将时间索引值向前推进了 900 天

这类迁移方法的常见使用场景就是计算数据在不同时段的差异。例如，我们可以用迁移后的值来计算 Google 股票一年期的投资回报率（如图所示）：

```python
ROI = 100 * (goog.shift(-365, 'D') / goog - 1)
ROI.plot()
plt.ylabel('% Return on Investment');
```

这可以帮助我们观察 Google 股票的总体特征：从图中可以看出，Google 的股票在 IPO 刚刚成功之后最值得投资（图里的趋势很直观），在 2009 年年中开始衰退。

3. 移动时间窗口

Pandas 处理时间序列数据的第 3 种操作是移动统计值（rollingstatistics）。这些指标可以通过 Series 和 DataFrame 的rolling() 属性来实现，它会返回与 groupby 操作类似的结果移动视图（rolling view）使得许多累计操作成为可能。例如，可以通过下面的代码获取 Google 股票收盘价的一年期移动平均值和标准差：

```python
rolling = goog.rolling(365, center=True)

data = pd.DataFrame({'input': goog,
                     'one-year rolling_mean': rolling.mean(),
                     'one-year rolling_std': rolling.std()})
ax = data.plot(style=['-', '--', ':'])
ax.lines[0].set_alpha(0.3)
```

## 13. 高性能Pandas: eval()与query()

Python 数据科学生态环境的强大力量建立在NumPy 与 Pandas 的基础之上，并通过直观的语法将基本操作转换成 C语言：在 NumPy 里是向量化 / 广播运算，在 Pandas 里是分组型的运算。虽然这些抽象功能可以简洁高效地解决许多问题，但是它们经常需要创建临时中间对象，这样就会占用大量的计算时间与内存。Pandas 从 0.13 版开始（2014 年 1 月）就引入了实验性工具，让用户可以直接运行 C 语言速度的操作，不需要十分费力地配置中间数组。它们就是 eval() 和 query() 函数，都依赖于 Numexpr 程序包。

### 13.1 eval()与query()的设计动机：复合代数式

Numpy与Pandas都支持快速的向量化运算，例如两个数组求和：

```python
import numpy as np
rng = np.random.RandomState(42)
x = rng.rand(1000000)
y = rng.rand(1000000)
%timeit x + y
```

这样做比普通的Python循环或列表综合要快很多：

```python
%timeit np.fromiter((xi + yi for xi, yi in zip(x, y)), dtype= x.dtype, count = len(x))
```

但是这种运算在处理复合代数式（compound expression）问题时的效率比较低，例如下面的表达式：

```python
mask = (x > 0.5) & (y < 0.5)

# 由于Numpy会计算每一个代数子式，因此该计算等价于
tmp1 = x > 0.5
tmp2 = y < 0.5
mask = tmp1 & tmp2
```

也就是说，每段中间过程都需要显式地分配内存。如果 x 数组和 y 数组非常大，这么运算就会占用大量的时间和内存消耗。Numexpr 程序库可以在不为中间过程分配全部内存的前提下，完成元素到元素的复合代数式运算。这么做的好处是计算比NumPy更高效，尤其适合处理大型数组。简单点说 Numexpr其实就是用一个 NumPy 风格的字符串代数式进行运算：

```python
import numexpr
mask_numexpr = numexpr.evaluate('(x > 0.5) & (y < 0.5)')
# allclose 函数通常用于数值数组，检查对应的元素对是否在某个容差范围内近似相等
np.allclose(mask, mask_numexpr)
```

### 13.2 用Pandas.eval()实现高性能运算：不同表间的运算

```python
import pandas as pd
norws, ncols = 10000, 100
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(norws, ncols)) for i in range(4))
```

如果用普通的Pandas方法计算四个DataFrame对象的和：

```python
%timeit df1 + df2 + df3 + df4
```

用pd.eval和字符串代数式计算得到相同的结果：

```python
%timeit pd.eval('df1 + df2 + df3 + df4')
```

pd.eval方法计算时间明显减少，而且内存消耗更少，结果也是一样的：

```python
np.allclose(df1 + df2 + df3 + df4, pd.eval('df1 + df2 + df3 + df4'))
```

pd.eval()支持许多运算，创建一个值在0-1000的100行3列的整数DataFrame演示：

```python
df1, df2, df3, df4, df5 = (pd.DataFrame(rng.randint(0, 1000, (100, 3)))
                           for i in range(5))
```

```python
# 1.算数运算符。pd.eval() 支持所有的算术运算符
result1 = -df1 * df2 / (df3 + df4) - df5
result2 = pd.eval('-df1 * df2 / (df3 + df4) - df5')
np.allclose(result1, result2)
```

```python
# 2.比较运算符。pd.eval() 支持所有的比较运算符，包括链式代数式(chained expression)
result1 = (df1 < df2) & (df2 <= df3) & (df3 != df4)
result2 = pd.eval('df1 < df2 <= df3 != df4')
np.allclose(result1, result2)
```

```python
# 3.位运算符。pd.eval() 支持 &（与）和|（或）等位运算符
result1 = (df1 < 0.5) & (df2 < 0.5) | (df3 < df4)
result2 = pd.eval('(df1 < 0.5) & (df2 < 0.5) | (df3 < df4)')
print(np.allclose(result1, result2))

# 另外还可以在布尔类型的代数式中使用 and 和 or 等字面值：
result3 = pd.eval('(df1 < 0.5) and (df2 < 0.5) or (df3 < df4)')
np.allclose(result1, result3)
```

```python
# 4.对象属性与索引。pd.eval() 可以通过 obj.attr语法获取对象属性，通过 obj[index] 语法获取对象索引：
result1 = df2.T[0] + df3.iloc[1]
result2 = pd.eval('df2.T[0] + df3.iloc[1]')
np.allclose(result1, result2)
```

目前 pd.eval() 还不支持函数调用、条件语句、循环以及更复杂的运算。如果想要进行这些运算，可以借助 Numexpr 来实现。

### 13.3 用DataFrame.eval()实现列间运算

由于pd.eval() 是 Pandas 的顶层函数，因此 DataFrame 有一个eval() 方法可以做类似的运算。使用 eval() 方法的好处是可以借助列名称进行运算

```python
df = pd.DataFrame(rng.rand(1000, 3), columns = ['A', 'B', 'C'])
df.head()
```

用pd.eval()进行三列间的计算(A + B) / (C - 1)：

```python
result1 = (df['A'] + df['B']) / (df['C'] - 1)
result2 = pd.eval("(df.A + df.B) / (df.C - 1)")
np.allclose(result1, result2)
```

而 DataFrame.eval() 方法可以通过列名称实现简洁的代数式：列名称作为变量来计算代数式结果同样正确

```python
result3 = df.eval('(A + B) / (C - 1)')
np.allclose(result1, result3)
```

1. DataFrame.eval()新增列

```python
df.head()
```

```python
# 用df.eval()创建一个新的列'D',然后赋给它其他列计算的值
df.eval('D = (A + B) / C', inplace = True) # inplace是否在原数据框中修改，默认为False
df.head()
```

```python
# 还可以修改已有的列
df.eval('D = (A - B) / C', inplace = True)
df.head()
```

2. DataFrame.eval()使用局部变量

```python
# 通过 @ 符号使用Python的局部变量
column_mean = df.mean(1)
result1 = df['A'] + column_mean
result2 = df.eval('A + @column_mean')
np.allclose(result1, result2)
```

@ 符号表示“这是一个变量名称而不是一个列名称”，从而让你灵活地用两个“命名空间”的资源（列名称的命名空间和 Python 对象的命名空间）计算代数式。需要注意的是，@ 符号只能在DataFrame.eval() 方法中使用而不能在 pandas.eval() 函数中使用，因为 pandas.eval() 函数只能获取一个（Python）命名空间的内容。

### 13.4 DataFrame.query()方法：对表内的筛选

```python
# 筛选同时满足A列小于0.5，B列小于0.5的行返回布尔系列作为索引，再返回对应的行
result1 = df[(df.A < 0.5) & (df.B < 0.5)]
result2 = pd.eval('df[(df.A < 0.5) & (df.B < 0.5)]')
result_dfeval = df[df.eval('A < 0.5 & B < 0.5')]
np.allclose(result1, result2)
np.allclose(result1, result_dfeval)
```

DataFrame.eval()语法不能像pd.eval(df['条件'])那样在函数内部直接使用df[]来直接输出满足条件的行，只能将df.eval('条件')返回的布尔值作为索引在函数外输入df[]中来输出满足条件的行，这种方法和df[(df.A < 0.5) & (df.B < 0.5)]一样都是显示的创建了布尔数组后再筛选，不如pd.eval高效节省内存。对于这种过滤运算就可以用 query() 方法：类似SQL的直接用列名比较的同时直接输出满足条件的行数据

```python
result2 = df.query('A < 0.5 and B < 0.5')
np.allclose(result1, result2)
```

除了计算性能更优之外，这种方法的语法也比掩码代数式语法更好理解。需要注意的是，query() 方法也支持用 @ 符号引用局部变量：

```python
Cmean = df['C'].mean()
result1 = df[(df.A < Cmean) & (df.B < Cmean)]
result2 = df.query('A < @Cmean and B < @Cmean')
np.allclose(result1, result2)
```

### 13.5 性能决定使用时机

在考虑要不要用这两个函数时，需要思考两个方面：计算时间和内存消耗，而内存消耗是更重要的影响因素。就像前面介绍的那样，每个涉及NumPy 数组或 Pandas 的 DataFrame 的复合代数式都会产生临时数组，例如：

```python
x = df[(df.A < 0.5) & (df.B < 0.5)]
```

它基本等价于：

```python
tmp1 = df.A < 0.5
tmp2 = df.B < 0.5
tmp3 = tmp1 & tmp2
x = df[tmp3]
```

如果临时 DataFrame 的内存需求比你的系统内存还大（通常是几吉字节），那么最好还是使用 eval() 和 query() 代数式。你可以通过下面的方法大概估算一下变量的内存消耗：

```python
df.values.nbytes
```

在性能方面，即使你没有使用最大的系统内存，eval() 的计算速度也比普通方法快。现在的性能瓶颈变成了临时 DataFrame 与系统 CPU 的L1 和 L2 缓存（在 2016 年依然是几兆字节）之间的对比了——如果系统缓存足够大，那么 eval() 就可以避免在不同缓存间缓慢地移动临时文件。在实际工作中，我发现普通的计算方法与 eval/query 计算方法在计算时间上的差异并非总是那么明显，普通方法在处理较小的数组时反而速度更快！eval/query 方法的优点主要是节省内存，有时语法也更加简洁。我们已经介绍了 eval() 与 query() 的绝大多数细节，若想了解更多的信息，请参考 Pandas 文档。尤其需要注意的是，可以通过设置不同的解析器和引擎来执行这些查询。

## 5. 数据提取

数据提取主要使用loc[]、iloc[]和ix三个函数，loc按标签值（索引标签）提取，iloc按位置提取，ix可以同时按标签和位置提取

### 5.1 按标签名称提取（loc-从0标记，因为是标签名称，故范围是左闭右闭区间）

标签指索引列的标签，注意df_inner.set_index('id')与df_inner = df_inner.set_index('id')的巨大差异
df_inner.set_index('id')是设置‘id’列为索引列，依然是默认索引[0:n],之后用loc函数提取数据且返回值与是否以其他列排序有关,但不改变数据表结构
df_inner = df_inner.set_index('id')不仅从原表抽出‘id’列作为索引列，还赋值了给原表，改变了数据表结构，原表已没有‘id’这个字段，数据提取不再用loc，而是直接用df_inner['id1':'id2'],且返回值与排序无关

```python
# 可见设置索引列时的排序已经定死了该列的索引标签，已其他字段排序时会跟着改动位置
df_inner = df_inner.sort_values(by = ['age'])
df_inner
```

```python
# 提取索引列为3的单行数据
df_inner.loc[3]
```

```python
# 提取索引列为0到5的数据行（不一定一定返回0,1,2,3,4,5行，与0到5的索引标签排序有关）
df_inner.loc[0:5]
```

```python
# reset_index函数用于恢复索引，重新将date字段的日期设置为数据表的索引，并按日期进行数据提取
#重设索引
df_inner.reset_index()
```

```python
# 提取2013-01-04之前的所有数据，没用loc，与排序无关
df_inner[:'2013-01-04']
```

可以按位置区域提取数据，也可以按位置逐条提取数据，冒号前后的数字不再代表标记名称，而是位置

Excel的数据目录下提供了“分类汇总”功能，可以按指定的字段和汇总方式对数据表进行汇总。Python中通过Groupby函数完成相应的操作，并可以支持多级分类汇总。Groupby是进行分类汇总的函数，使用方法很简单，制定要分组的列名称就可以，也可以同时制定多个列名称，groupby按列名称出现的顺序进行分组。同时要制定分组后的汇总方式，常见的是计数和求和两种。
